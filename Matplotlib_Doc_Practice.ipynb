```python
def math_plot(x, y):
        mlt.plot(x,y, label='linear')
        mlt.plot(x,y**2, label='quadratic')
        mlt.plot(x,y**3, label='cubic')
        mlt.plot(x,np.exp(y*0.6), label='exp')
        mlt.plot(x,np.sin(y*6), label='sin')
        mlt.plot(x,np.cos(y*6), label='cos')
        # mlt.plot(x,np.tan(y**0.6), label='tan')
        mlt.xlabel("0 to 2 range", size=16)
        mlt.ylabel("Mathematical Output", size=16)
        mlt.legend()
        mlt.title("Mathematical Plot", color='grey', size=20)
        mlt.show()
```

```python
a = np.linspace(0,1,100)
b = np.linspace(0,2,100)
# print(a,"\n"*3, b)

math_plot(a,b)
```

```python
def ax_plot(ax, data1, data2, **parameter_dictionary):
    out = ax.plot(data1, data2, **parameter_dictionary)
    return out
```

```python
fig, ax = mlt.subplots(figsize=(10,5))

x = np.arange(len(d1))
ax.plot(x, np.cumsum(d1), color='red', linewidth=3.3, linestyle='--')

l2, = ax.plot(x, np.cumsum(d2), color='red', linewidth=2)
l2.set_linestyle(':')

l3, = ax.plot(x, np.cumsum(d3), color='red', linewidth=2, label='L3')
l3.set_linestyle(':')
```

```python
fig, ax = mlt.subplots(figsize=(10,5), layout='tight')

lines = ax.plot(x, np.cumsum(d1), x, np.cumsum(d2), x, np.cumsum(d3), x, np.cumsum(d4), linewidth=1.6, linestyle='--')

lines[0].set_color('grey')
lines[0].set_label('Line d1')

lines[1].set_color('red')
lines[1].set_label('Line d2')

lines[2].set_color('green')
lines[2].set_label('Line d3')

lines[3].set_color('blue')
lines[3].set_label('Line d4')

ax.legend()

mlt.show()
```

```python
from matplotlib.colors import LogNorm

x, y = np.meshgrid(np.linspace(-3,3,128), np.linspace(-3,3,128))
z = (1 - x/2 + x**5 + y**3)*np.exp(-x**2-y**2)
```

```python
fig, axs = mlt.subplots(2,2, layout='constrained')

pc = axs[0,0].pcolormesh(x, y, z, vmin= -1, vmax=1, cmap='RdBu_r')
fig.colorbar(pc, ax=axs[0,0])
axs[0,0].set_title('pcolormesh()')

co = axs[0,1].contourf(x, y, z, levels=np.linspace(-1.25, 1.25, 11))
fig.colorbar(co, ax=axs[0,1])
axs[0,1].set_title('contourf()')

pc = axs[1,0].imshow(z**2*100, cmap='plasma', norm=LogNorm(vmin=0.01, vmax=100))
fig.colorbar(pc, ax=axs[1,0])
axs[1,0].set_title('imshow() with LogNorm()')

pc = axs[1,1].scatter(x, y, c=(x+y), cmap='RdBu_r')
fig.colorbar(pc, ax=axs[1,1], extend='both')
axs[1,1].set_title('scatter()')

```
