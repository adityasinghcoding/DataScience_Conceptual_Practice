```python
import matplotlib.pyplot as plt
```

```python
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(10,5))

sns.swarmplot(data=tips, x='sex', y='total_bill', hue='tip', ax=ax1)
ax1.set_title('Gender vs Total Bill with TIP comparison', weight='heavy')

sns.swarmplot(data=tips, x='sex', y='total_bill', hue='total_bill', ax=ax2)
ax2.set_title('Gender vs Total Bill with Total Bill intensity', weight='heavy')

plt.tight_layout()
plt.show()
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
sns.jointplot(data=penguins, x='flipper_length_mm', y='bill_length_mm', hue='species')
```

```python
sns.pairplot(data=penguins, hue='species')
```

```python
g = sns.PairGrid(penguins, hue='species', corner=False)
```

```python
g = sns.PairGrid(penguins, hue='species', corner=False)
g.map_lower(sns.kdeplot, hue=None, levels=5, color='.3')
```

```python
g = sns.PairGrid(penguins, hue='species', corner=False)
g.map_lower(sns.kdeplot, hue=None, levels=5, color='.1')
g.map_lower(sns.scatterplot, marker='*')
```

```python
g = sns.PairGrid(penguins, hue='species', corner=False)
g.map_lower(sns.kdeplot, hue=None, levels=5, color='.1')
g.map_lower(sns.scatterplot, marker='*')
g.map_diag(sns.histplot, element='step', linewidth=2, kde=True)
```

```python
g = sns.PairGrid(penguins, hue='species', corner=True)
g.map_lower(sns.kdeplot, hue=None, levels=5, color='.1')
g.map_lower(sns.scatterplot, marker='*')
g.map_diag(sns.histplot, element='step', linewidth=2, kde=True)
g.add_legend(frameon=True)
g.legend.set_bbox_to_anchor((.61, .6))
```

```python
sns.set_theme(style='ticks', font_scale=1.5)
g = sns.relplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', 
                hue='body_mass_g', palette='crest', marker='x', s=100)

g.set_axis_labels('Bill length(mm)', 'Bill depth(mm)', labelpad=6)
g.legend.set_title('Body mass(g)')
g.figure.set_size_inches(10,5)
g.ax.margins(.16)
g.despine(trim=True)
```
